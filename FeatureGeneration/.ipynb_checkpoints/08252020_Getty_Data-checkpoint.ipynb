```python
def create_features(data):
    hate = HateWrapper()
    sentiment_classifier = TextClassifier.load('en-sentiment')
    scores = []

    hate_scores = hate.predict(data)
    for el in data:
        new_data = {}
        new_data['hate_speech'] = hate_scores[hate_scores.text == el].hate_speech
        new_data['offensive_language'] = hate_scores[hate_scores.text == el].offensive_language
        new_data['neither'] = hate_scores[hate_scores.text == el].neither

        sentence = Sentence(el)
        sentiment_classifier.predict(sentence)
        label = str(sentence.labels[0]).split(' ')[0]
        proba = float(str(sentence.labels[0]).split(' ')[1].replace('(', '').replace(')', ''))
        new_data['sentiment'] = [label, proba]
        scores.append(new_data)
        
    return scores
```
